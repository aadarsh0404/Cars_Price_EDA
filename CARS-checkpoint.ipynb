```python
outlier_reason = []
ct = 0
for i in new_cars['price'].values:
    ct = ct + 1
#print(ct)
for i in range(0,ct):
    outlier_reason.append('NA')
for col in new_cars.columns:
    c = 0
    q3 = new_cars[col].quantile(0.75)
    q1 = new_cars[col].quantile(0.25)
    iqr = q3 - q1
    maximum = q3 + 1.5*iqr
    minimum = q1 - 1.5*iqr
    #print('range iqr',col,minimum,'to ',maximum)
    for i in new_cars[col].values:
        if(i>maximum or i<minimum):
            outlier_reason[c] = 'Due to '+col+ ' in iqr'
            c = c + 1
        else:
            c = c + 1
outlier_reason
```

```python
for col in new_cars.columns:
    c = 0
    M = max(new_cars[col])
    m, s = np.mean(new_cars[col]), np.std(new_cars[col])
    L, H = m-1.6667*s, m+1.6667*s
    #plt.subplots()
    #print('range sd',col,L,'to ',H)
    #sns.distplot(new_cars[col])
    for i in new_cars[col].values:
        if(i<L or i>H):
            if('iqr' in outlier_reason[c]):
                outlier_reason[c] = 'Due to '+col+ ' in iqr and sd'
            else:
                outlier_reason[c] = 'Due to '+col+ ' in sd'
            c = c + 1
        else:
            c = c + 1
new_cars['outlier reason'] = outlier_reason
```
