```python
#Now we make 2 copies of dataset and one, we replace 0 with NaN(Except for symboling) and for other we visualize mean, Median/Mode replacement
dataset=data.copy()
dataset1=data.copy()
# for i in num_feat:
#     sns.distplot(data[i])
#     plt.show()
    
for i in num_feat:
    if i != 'symboling':
        dataset[i] = np.where(dataset[i]==0,dataset[i].mean(),dataset[i])
        dataset1[i] = np.where(dataset1[i]==0,np.NaN,dataset[i])

%matplotlib inline 
for i in num_feat:
    fig,ax = plt.subplots()
    sns.distplot(dataset[i],label='mean replaced')
    sns.distplot(dataset1[i].dropna(),label='NaN removed')
    plt.title(i)
    ax.legend()
```

```python
#We will implement median replacement due to presence of outliers
for i in num_feat:
    if i != 'symboling':
        data[i] = np.where(data[i]==0,data[i].median(),data[i])
```

```python
def interpolation(X):
    Y = ((27)*(X-1))+68
    return Y
def revinter(Y):
    X = ((Y-68)/27)+1
    return X
```

```python
revinter(90)
# interpolation(1.5)
```

```python
values1 = []
c = 0
ct = len(dataset.price)
for i in range(0,ct):
    values1.append(False)
    
for col in num_feat:
    c = 0
    M = max(dataset[col])
    m, s = dataset[col].mean(), np.std(dataset[col])
    L, H = m-revinter(90)*s, m+revinter(90)*s
    plt.show()
    print('range sd',col,L,'to ',H)
    sns.distplot(dataset[col])
    for i in dataset[col].values:
        if(i<L or i>H):
            values1[c]= True
            c = c + 1
        else:
            c = c + 1
print(ct)
```

```python
values = []
ct = len(dataset.price)
print(ct)
for i in range(0,ct):
    values.append(False)
for col in num_feat:
    c = 0
    q3 = dataset[col].quantile(0.75)
    q1 = dataset[col].quantile(0.25)
    iqr = q3 - q1
    maximum = q3 + 1.5*iqr
    minimum = q1 - 1.5*iqr
    print('range iqr',col,minimum,'to ',maximum)
    for i in dataset[col].values:
        if(i>maximum or i<minimum):
            values[c] = True
            c = c + 1
        else:
            c = c + 1
    plt.subplots()
    sns.distplot(dataset[col])
```

```python
outlier_reason = []
ct = 205

for i in range(0,ct):
    outlier_reason.append('NA')
for col in num_feat:
    c = 0
    q3 = dataset[col].quantile(0.75)
    q1 = dataset[col].quantile(0.25)
    iqr = q3 - q1
    maximum = q3 + 1.5*iqr
    minimum = q1 - 1.5*iqr
    #print('range iqr',col,minimum,'to ',maximum)
    for i in dataset[col].values:
        if(i>maximum or i<minimum):
            if 'iqr' not in outlier_reason[c]:
                outlier_reason[c] = 'In iqr '+'due to '+col
            else:
                outlier_reason[c] = outlier_reason[c]+' and '+col
            c = c + 1
        else:
            c = c + 1
outlier_reason
```

```python
for col in num_feat:
    c = 0
    M = max(dataset[col])
    m, s = np.mean(dataset[col]), np.std(dataset[col])
    L, H = m-1.6667*s, m+1.6667*s
    for i in dataset[col].values:
        if(i<L or i>H):
            if('iqr' in outlier_reason[c]):
                if('sd' not in outlier_reason[c]):
                    outlier_reason[c] = 'In iqr and sd'+' due to '+col
                else:
                    outlier_reason[c] = outlier_reason[c]+' and '+col
            else:
                if('sd' not in outlier_reason[c]):
                    outlier_reason[c] = 'In sd'+' due to '+col
                else:
                    outlier_reason[c] = outlier_reason[c]+' and '+col
            c = c + 1
        else:
            c = c + 1
outlier_reason
```
